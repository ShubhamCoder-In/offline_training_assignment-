```python
population_india = [1021,1200,2100,1800,1900,2400,4300,6000]
population_china = [1500,1800,2400,2100,1900,2600,3300,4000]
population_us = [1300,1500,2600,2100,1900,2100,3300,2900]
year =  [2012,2013,2014,2015,2016,2017,2018,2019]
pit.plot(year, population, marker="*", color="red",ms=10,mfc="yellow",lw=4, ls="dashed")
pit.title("population trend")
pit.xlabel("year")
pit.ylabel("population")
pit.show()
```

```python
population_india = [1021,1200,2100,1800,1900,2400,4300,6000]
population_china = [1500,1800,2400,2100,1900,2600,3300,4000]
population_us = [1300,1500,2600,2100,1900,2100,3300,2900]
year =  [2012,2013,2014,2015,2016,2017,2018,2019]
pit.plot(year, population_india, marker="*", color="red",ms=10,mfc="yellow",lw=4, ls="dashed",label="india")
pit.plot(year, population_china, marker="*", color="yellow",ms=10,mfc="yellow",lw=4, ls="dotted",label="china")
pit.plot(year, population_us, marker="*", color="orange",ms=10,mfc="yellow",lw=4,label="us")
pit.legend
pit.title("population trend")
pit.xlabel("year")
pit.ylabel("population")
pit.show()
```

```python
ownership = list(df.owner.value_counts().keys())
no_of_counts = list(df.owner.value_counts().values)
pit.scatter(ownership,no_of_counts)
```

```python
ownership = list(df.owner.value_counts().keys())
no_of_counts = list(df.owner.value_counts().values)
pit.bar(ownership,no_of_counts, color="green")
pit.xticks(rotation=60)
```

```python
ownership = list(df.brand.value_counts().keys())
no_of_counts = list(df.brand.value_counts().values)
pit.bar(ownership,no_of_counts, color="green", ec="black")
pit.xticks(rotation=90)
```
