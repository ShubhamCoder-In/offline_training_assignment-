```python
df3 = pd.get_dummies(df,drop_first="if_binary").astype(int)
df3.head(3)
```

```python
df3.drop('loan_id',axis=1,inplace=True)
```

```python
x = df3.drop('loan_status_Rejected',axis=1)
y = df3[['loan_status_Rejected']]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std = StandardScaler()
```

```python
x=std.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Sequential
```

```python
x_train.shape
```

```python
ann_model = Sequential()
ann_model.add(Dense(units=68,activation="relu",input_dim=11, name='hiden1'))
ann_model.add(Dense(units=35,activation='relu',name='hiden2'))
ann_model.add(Dense(units=15,activation='relu',name='hiden3'))
ann_model.add(Dense(units=1,activation='sigmoid',name='outpulayer'))
ann_model.compile(loss='binary_crossentropy',optimizer='adam', metrics=['accuracy'])
```

```python
ann_model.summary()
```

```python
history = ann_model.fit(x_train,y_train,epochs=5,validation_data=(x_test,y_test))
```

```python
history = pd.DataFrame(history)
history
```

```python
history.plot()
```

```python
prediction = ann_model.predict(x_test)
```

```python
prediction = (prediction>0.5).astype(int).ravel()
```

```python
y_test,
```

```python
from sklearn.metrics import confusion_matrix
```

```python
metrix = confusion_matrix(y_test,prediction)
metrix
```
