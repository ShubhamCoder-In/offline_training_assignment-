```python
from sklearn.preprocessing import StandardScaler
```

```python
sd = StandardScaler()
```

```python
x_transfrom = sd.fit_transform(X)
```

```python
pd.DataFrame(x_transfrom).describe()[0][1]
```

```python
desc = pd.DataFrame(x_transfrom).describe()
desc_formatted = desc.applymap(lambda x: f"{int(x):d}")
desc_formatted
```

```python
from sklearn.cluster import KMeans
```

```python
Km = KMeans()
```

```python
Km.fit(x_transfrom)
```

```python
label = Km.labels_
```

```python
pd.DataFrame(x_transfrom,columns=X.columns)
```

```python
X['cluster_8']=label
```

```python
X.head(20)
```

```python
df['cluster_8'] = label
```

```python
df[df['label']=='rice']['cluster_8'].value_counts()
```

```python
df[df['cluster_8']==1]['label'].value_counts()
```

```python
Km.score(x_transfrom,label)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_score(x_transfrom,label)
```

```python
Km.inertia_ #wcss ( with in )
```

```python
wcss = []
for i in range(5,51):
    Km1= KMeans(n_clusters=i)
    Km1.fit(x_transfrom)
    wcss.append(Km1.inertia_)
```

```python
plt.title('Elbow method')
plt.plot(list(range(5,51)),wcss)
plt.axvline(12)
plt.show()
```

```python
Km3 = KMeans(n_clusters=12)
```

```python
Km3.fit(x_transfrom)
```

```python
label2 = Km3.labels_
```

```python
silhouette_score(x_transfrom,label2)
```
